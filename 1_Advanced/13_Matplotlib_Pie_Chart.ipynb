```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
import ast
# Loading dataset
data_set = load_dataset('lukebarousse/data_jobs')
df = data_set['train'].to_pandas()
# Cleaning dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda skill_list : ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
plt.pie(df['job_work_from_home'].value_counts(),autopct='%1.1f%%',startangle=90,labels=['False','True'])
plt.title('Is Work From Home? (%)')
```

```python
dict = {
    'job_work_from_home' : 'is work from Home ? (%)',
    'job_no_degree_mention' : 'is degree Mentioned ? (%)',
    'job_health_insurance' : 'Health Insurance (%)'
}
fig,ax = plt.subplots(1,3)
for i,(column,title) in enumerate(dict.items()):
    ax[i].pie(df[column].value_counts(),startangle=90,labels=['False','True'],autopct='%1.1f%%')
    ax[i].set_title(title)
    fig.tight_layout()
```
