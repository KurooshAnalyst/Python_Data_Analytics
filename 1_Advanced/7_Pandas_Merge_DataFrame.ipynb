```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
# Loading dataset
data_set = load_dataset('lukebarousse/data_jobs')
df = data_set['train'].to_pandas()
df_us_software = pd.read_csv('software_jobs.csv',index_col='job_posted_month')
# Cleaning dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] =='United States'].copy()
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')
df_us_pivot = df_us.pivot_table(index='job_posted_month',columns='job_title_short',aggfunc='size')
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month_no'] = pd.to_datetime(df_us_pivot['job_posted_month'],format='%B')
df_us_pivot['month_no'] = df_us_pivot['month_no'].dt.month
df_us_pivot.sort_values(by='month_no',inplace=True)
df_us_pivot.set_index(keys='job_posted_month',inplace=True)
df_us_pivot.drop(columns='month_no')

```

```python
df_us_merge = df_us_pivot.merge(df_us_software,on='job_posted_month')
df_us_merge.drop('month_no',axis=1,inplace=True)
```

```python
top_5 = (df_us_merge.sum()
         .sort_values(ascending=False)
         .head()
         .index
         .tolist())
```

```python
df_us_merge[top_5].plot(kind='line')
plt.ylabel('Job Count')
plt.xlabel('2023')
plt.ylim(0,20000)
plt.title('High Demand Data and Software Jobs')
plt.show()
```
