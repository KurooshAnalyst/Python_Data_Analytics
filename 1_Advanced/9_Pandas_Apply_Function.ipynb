```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
import ast
# Loading dataset
data_set = load_dataset('lukebarousse/data_jobs')
df = data_set['train'].to_pandas()
# Cleaning dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])

```

```python
df_salary = df[pd.notna(df['salary_year_avg'])].copy()
df_salary.head()
```

```python
df_salary['salary_year_inflacted'] = df_salary['salary_year_avg'].apply(lambda salary : salary * 1.03)
```

```python
df_salary[['salary_year_avg','salary_year_inflacted']].head()
```

```python
sample_skill = df['job_skills'][1]
sample_skill = ast.literal_eval(sample_skill)
sample_skill
```

```python
df_altered = df['job_skills'].apply(lambda skill_list : ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
df_altered
```

```python
print('Old Type Of Skills : ', type(df['job_skills'][2]))
print('After Manipulation : ', type(df_altered[2]))
```

```python
df_salary_expert = df[pd.notna(df['salary_year_avg'])].copy()
```

```python
def projected_salary(salary):
    if "Senior" in salary['job_title_short']:
        return salary['salary_year_avg'] * 1.05
    else :
        return salary['salary_year_avg'] * 1.03
 
df_salary_expert['salary_year_inflated'] = df_salary_expert.apply(projected_salary , axis=1)
df_salary_expert[['job_title_short','salary_year_avg','salary_year_inflated']]
```
