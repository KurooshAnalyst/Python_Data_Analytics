```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset
# Loading dataset
data_set = load_dataset('lukebarousse/data_jobs')
df = data_set['train'].to_pandas()
# Cleaning dataset
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.strftime('%b')
df['job_posted_month']
```

```python
months = df['job_posted_month'].unique()
```

```python
dict_df = {month: df[df['job_posted_month'] == month] for month in months}
```

```python
df_Q1 = pd.concat([dict_df['Jan'],dict_df['Feb'],dict_df['Mar']],ignore_index=True)
```

```python
df_Q1['job_posted_month'].value_counts().plot(kind='bar')
plt.title('How was Job Postings On First Quarter Of 2023')
plt.ylabel('Job Posting Count')
plt.xlabel('Quarter 1')
plt.xticks(rotation = 45 , ha= 'right')
plt.show()
```
